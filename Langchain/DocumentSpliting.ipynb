# **<span style="color: green;">Best Way to Split!</span>**

*<span style="background-color: yellow; color: red; padding: 2px;"> While loading pdf using pdfloader, it's breaks down into the pages, page break can affect the text splitting process, hence before splitting some text processing is required: get list of extracted pdf pages -> combine it together -> then split</span>*

```python
from langchain.document_loaders import PyMuPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Load the PDF using PyMuPDFLoader
def load_pdf(pdf_path):
    loader = PyMuPDFLoader(pdf_path)
    documents = loader.load()  # This loads the PDF as a list of Document objects
    # Extract text from the documents
    text = "\n".join([doc.page_content for doc in documents])
    return text

# Clean the text (optional step)
def clean_text(text):
    # Remove unwanted line breaks, extra spaces, etc.
    text = ' '.join(text.split())  # Collapse multiple spaces into one
    return text

# Split the text into chunks with overlap
def split_text_with_overlap(text, chunk_size, chunk_overlap):
    splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
    return splitter.split_text(text)

# Main function to process the PDF
def process_pdf(pdf_path, chunk_size, chunk_overlap):
    # Load and clean the text from the PDF
    text = load_pdf(pdf_path)
    cleaned_text = clean_text(text)
    print("Cleaned tet: ", cleaned_text)
    print("*"*50, end="\n"*2)
    
    # Split the cleaned text into chunks with overlap
    chunks = split_text_with_overlap(cleaned_text, chunk_size, chunk_overlap)
    splitted_data=[]
    # Print out the chunks to check overlap
    for i, chunk in enumerate(chunks):
        # print(f"Chunk {i + 1}: {chunk[:200]}...")  # Print first 200 characters of each chunk
        print(f"Chunk {i + 1}: {chunk}")
        splitted_data.append(chunk)
    return splitted_data

# Example usage
pdf_path = "./Documents/Attenton all you need.pdf"  # Path to your PDF file
chunk_size = 500  # Size of each chunk (number of characters)
chunk_overlap = 50  # Overlap between chunks (number of characters)

# Process the PDF
ans = process_pdf(pdf_path, chunk_size, chunk_overlap)
ans
```

```python
len(ans)
```

```python
# to debug
# Check overlap between adjacent chunks
for i in range(1, len(ans)):
    print(f"End of chunk {i}: {ans[i-1][-50:]}")  # Last 50 chars of previous chunk
    print(f"Start of chunk {i+1}: {ans[i][:50]}")  # First 50 chars of current chunk
    print("-" * 50)
```

# **Conclusion:**

Use PyMuPDFLoader to load and extract text from your PDF, and ensure the text is cleaned up before splitting.

Use RecursiveCharacterTextSplitter to split the text into chunks with overlap.
